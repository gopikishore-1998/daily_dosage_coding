```python
import pandas as pd
import numpy as np
```

```python
#Problem: Create a NumPy array and compute mean, median, variance.
arr = np.array([10,30,25,30,40,80,50,70,200,167,80])
print(arr)

# Calculate the mean
print(np.mean(arr))
print(arr.mean())
# Calculate the median
print(arr.var())
print(np.var(arr))
print(np.median(arr))
print(pd.Series(arr).median())
```
